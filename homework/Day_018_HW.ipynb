```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = '../Part01//'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace(365243, np.nan, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

# 原始數據

```python
app_train['DAYS_BIRTH_YEARS'] = app_train['DAYS_BIRTH']/365
```

```python
app_train['DAYS_BIRTH_YEARS'].describe()
```

```python
app_train.groupby('TARGET')['DAYS_BIRTH_YEARS'].mean()
```

```python
app_train['equal_width_BIRTH_YEARS'] = pd.cut(app_train['DAYS_BIRTH_YEARS'], 10)
app_train['equal_freq_BIRTH_YEARS'] = pd.qcut(app_train['DAYS_BIRTH_YEARS'], 5)
DBY_0 = app_train.loc[app_train['TARGET']==0, :]
DBY_1 = app_train.loc[app_train['TARGET']==1, :]
```

```python
app_train.boxplot(column='DAYS_BIRTH_YEARS', by='TARGET', figsize=[8,5])
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=[8,5])
sns.distplot(DBY_0['DAYS_BIRTH_YEARS'], label = 'Target_0', color='red')
sns.distplot(DBY_1['DAYS_BIRTH_YEARS'], label = 'Target_1', color='purple')
plt.legend()
plt.title('KDE plot of DAYS_BIRTH_YEARS')
plt.xlabel('years')
plt.show()
```
