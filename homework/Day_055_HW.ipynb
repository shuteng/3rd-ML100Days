# 作業

- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
%matplotlib inline
```

```python
pca_X = PCA(n_components=2).fit_transform(X)
```

```python
plt.scatter(pca_X[:,0],pca_X[:,1], c=y)
plt.title('clusters with label')
```

```python
plt.figure(figsize=[10,8])
for i, j in zip(np.arange(1,5), np.arange(1,8,2)):
    km = KMeans(n_clusters=j).fit(X)
    labels = km.labels_
    plt.subplot(2,2,i)
    plt.scatter(pca_X[:,0],pca_X[:,1], c=labels)
    plt.title(f'kmeans clusters = {j}')
    plt.xticks([])
    plt.yticks([])
```

-----------------

3d projection

```python
pca_X3d = PCA(n_components=3).fit_transform(X)
```

```python
plt.figure(figsize=[6,4])
ax = plt.axes(projection='3d')
ax.scatter3D(pca_X3d[:,0],pca_X3d[:,1],pca_X3d[:,2], c=y)
plt.title('clusters with label 3d')
```

```python
km = KMeans(n_clusters=4, random_state=1).fit(X)
labels = km.labels_
```

```python
plt.figure(figsize=[6,4])
ax = plt.axes(projection='3d')
ax.scatter3D(pca_X3d[:,0],pca_X3d[:,1],pca_X3d[:,2], c=labels)
plt.title('kmeans clusters 3d')
```

Random init

```python
km = KMeans(n_clusters=4, random_state=1, init='random', n_init=1).fit(X)
labels = km.labels_
```

```python
plt.figure(figsize=[6,4])
ax = plt.axes(projection='3d')
ax.scatter3D(pca_X3d[:,0],pca_X3d[:,1],pca_X3d[:,2], c=labels)
plt.title('kmeans clusters 3d (random init)')
```
